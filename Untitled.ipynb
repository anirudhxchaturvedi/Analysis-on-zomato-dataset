```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head
```

```python
#we can combine the country column in both data frames using merge 
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
final_df.columns
```

```python
final_df.Country.value_counts() 
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
##pie chart
plt.pie(country_val,labels=country_names)
```

#### but the above chart looks really bad so we consider some top values may be top 3 or top 5

```python
#for top 3 
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

observation:Zomato maximum records or transaction are from India after that usa and then united kingdom.

```python
final_df.columns
```

```python
#we'll try to relate rating using aggregate rating, it's color and text 
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

## Observation
1. when the rating in between 4.5 to 4.9->excellent
2. when the rating in between 4.0 to 4.4->very good
3. when the rating in between 3.5 to 3.9->good
4. when the rating in between 3.0 to 3.4->average
5. when the rating in between 2.5 to 2.9->average
6. when the rating in between 2.0 to 2.4->poor

```python
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
#now, this graph is not clear and big enough to read the data carefully
#so we use matplotlib to resize the the bar chart
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

### Observation 
1. not rated count is very high
2. maximum number of rating are between 2.5 to 3.4

```python
## count plot  
sns.countplot(x="Rating color",data=ratings,palette=['black','red','orange','yellow','green','green'])
```

```python
## find out which currency is used by which country 
final_df.columns
```

```python
final_df[['Country','Currency']].groupby(['Country','Currency']).size().reset_index()
```

1. Group by 'Country' and 'Currency': Groups the data based on unique combinations of 'Country' and 'Currency'.
2. Count the size of each group: Counts the number of occurrences in each group.
3. Reset the index: Converts the group labels into columns and resets the index to a default integer index.

```python
## which countries do have online deliveries
final_df[final_df['Has Online delivery']=="Yes"].Country.value_counts()
```

```python
final_df[['Has Online delivery','Country']].groupby(['Has Online delivery','Country']).size().reset_index()
```

So, the online delivery is available only in INDIA and UAE
